##**Analisis de Correlacion**

###**Objetivo**
Explotar cómo se relaionan varabes numericas como duración, promedio de votos,  votos totales,  etc. para identificar posibles patrones entre popularidad y calidad



```python
import seaborn as sns
import matplotlib.pyplot as plt

#Selecionamos solo columnas numericas relevntes
columnas_correlación = [
    'duration', 'avg_vote', 'votes', 'weighted_average_vote', 'total_votes'
]
# nos aeguramos que no hay nulos
df_corr=df_merged[columnas_correlación].dropna()

#Calculamos la matriz de correlación
matrizz_correlación=df_corr.corr()
print(matrizz_correlación)

#Creamos el heatmap
plt.figure(figsize=(10,8))
sns.heatmap(matrizz_correlación,annot=True,cmap='coolwarm',fmt='.2f',linewidths=.5)
plt.title('Matriz de correlaion ente variables')
plt.show()
```

###Como notas de lo que podemos observar en el heatmap
identificamos que:
  * La duración tiene baja correlación con la calificación promedio
  * Las variables de votos (**votes**, **total_votes**) están altamente correlacionadas, lo cual tiene sentido (más votos = más exposición).
  * La **avg_vote** y la **weighted_average_vote** tienen correlación muy fuerte (> 0.9), indicando que los sistemas de calificación están alineados.

```python
#haremos una grafica scatterplot
sns.scatterplot(data=df_corr, x='votes', y='avg_vote')
plt.title('Relación entre votos y calificación promedio')
plt.xlabel('Votos')
plt.ylabel('Calificación promedio')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Suponiendo que ya tienes tu df_genres con las columnas:
# ['title', 'year', 'genre', 'votes', ...]
# Asegúrate de que 'year' y 'votes' sean numéricos:
df_merged['year'] = pd.to_numeric(df_merged['year'], errors='coerce')
df_merged['votes'] = pd.to_numeric(df_merged['votes'], errors='coerce')

# Selecciona los géneros que quieres graficar (por ejemplo, los 9 más frecuentes)
top_genres = (df_merged['genre']
              .value_counts()
              .nlargest(9)
              .index
             ).tolist()

df_top = df_merged[df_merged['genre'].isin(top_genres)]
years = sorted(df_top['year'].dropna().unique())

# Preparamos grid de subplots
n = len(top_genres)
cols = 3
rows = int(np.ceil(n / cols))
fig, axes = plt.subplots(rows, cols, figsize=(16, 5 * rows), sharey=True)

# Colores para cada género
cmap = plt.get_cmap('tab10')
colors = cmap(np.linspace(0, 1, n))

for ax, genre, color in zip(axes.flatten(), top_genres, colors):
    data = [df_top[(df_top['genre'] == genre) & (df_top['year'] == y)]['votes']
            for y in years]
    # Dibujamos el boxplot
    b = ax.boxplot(
        data,
        positions=range(len(years)),
        patch_artist=True,
        widths=0.6,
        showfliers=False
    )
    # Asignamos color a cada caja
    for patch in b['boxes']:
        patch.set_facecolor(color)
        patch.set_edgecolor('black')

    ax.set_title(genre, fontsize=12)
    ax.set_xticks(range(len(years)))
    ax.set_xticklabels(years, rotation=45, fontsize=8)
    ax.set_xlabel('Año')
    ax.grid(axis='y', linestyle='--', alpha=0.5)

# Quitamos ejes vacíos si los hay
for ax in axes.flatten()[n:]:
    ax.set_visible(False)

fig.suptitle('Distribución de Popularidad (Votos) por Año y Género', fontsize=16, y=1.02)
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Asegúrate de que las columnas son numéricas
df_merged['year'] = pd.to_numeric(df_merged['year'], errors='coerce')
df_merged['votes'] = pd.to_numeric(df_merged['votes'], errors='coerce')
df_merged['avg_vote'] = pd.to_numeric(df_merged['avg_vote'], errors='coerce')
df_merged['duration'] = pd.to_numeric(df_merged['duration'], errors='coerce')

# Selecciona las features numéricas que quieres comparar contra popularidad
features = ['year', 'duration', 'avg_vote', 'votes']

# Renombra para claridad en ejes
df_plot = df_merged[features].rename(columns={
    'votes': 'popularidad',
    'avg_vote': 'calidad_percibida',
    'duration': 'duración'
})

# Grafica pairplot con regresión y paleta de colores
sns.set(style="whitegrid")
g = sns.pairplot(
    df_plot,
    x_vars=[f for f in df_plot.columns if f != 'popularidad'],
    y_vars='popularidad',
    kind='reg',
    plot_kws={'line_kws':{'color':'gold'}, 'scatter_kws':{'alpha':0.3}},
    palette='tab10',
    height=3.5
)

g.fig.suptitle("Popularidad vs Características Numéricas de Películas", y=1.02)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Asegurar que el año sea numérico
df_merged['year'] = pd.to_numeric(df_merged['year'], errors='coerce')

# Filtrar películas desde el año 2000
df_merged = df_merged[df_merged['year'] >= 2000]

# Separar géneros múltiples y limpiar espacios
df_merged = df_merged.assign(genre=df_merged['genre'].str.split(',')).explode('genre')
df_merged['genre'] = df_merged['genre'].str.strip()

# Obtener los géneros más populares (top 10)
top_genres = df_merged['genre'].value_counts().nlargest(10).index
df_merged = df_merged[df_merged['genre'].isin(top_genres)]

# Crear rangos de años
bins = [2000, 2005, 2010, 2015, 2020, 2025]
labels = ['2000-2005', '2006-2010', '2011-2015', '2016-2020', '2021-2025']
df_merged['year_range'] = pd.cut(df_merged['year'], bins=bins, labels=labels)

# Agrupar datos
grouped = df_merged.groupby(['genre', 'year_range']).size().reset_index(name='count')

# Crear gráfico tipo aluvial con estilo mejorado
fig = px.parallel_categories(
    grouped,
    dimensions=['genre', 'year_range'],
    color='count',
    color_continuous_scale='Tealrose',  # Paleta de colores más suave y moderna
    labels={'genre': 'Género', 'year_range': 'Rango de Año'},
)

# Mejorar el diseño del gráfico
fig.update_layout(
    title={
        'text': 'Evolución de Géneros de Películas Populares (2000-2025)',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    title_font_size=22,
    font=dict(
        family="Arial, sans-serif",
        size=14,
        color="black"
    ),
    coloraxis_colorbar=dict(
        title="Cantidad",
        titlefont_size=14,
        tickfont_size=12,
    ),
    margin=dict(l=40, r=40, t=80, b=40)
)

fig.show()
```
